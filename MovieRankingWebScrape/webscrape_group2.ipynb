```python
tg_container = grosstable_table[0]
print (tg_container.find(class_='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank').get_text())
print (tg_container.find(class_='a-text-left mojo-field-type-title').get_text())
print (tg_container.find(class_='a-text-right mojo-field-type-money').get_text())
print (tg_container.find(class_='a-text-right mojo-field-type-percent').get_text())
```

```python
movie_ranks=[]
r1 = soup.find_all(class_='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank')
for r in r1:
    movie_ranks.append(r.text)
print (movie_ranks)
```

```python
movie_titles=[]
t1 = soup.find_all(class_='a-text-left mojo-field-type-title')
for t in t1:
    movie_titles.append(t.text)
print (movie_titles)
```

```python
movie_gross=[]
gr1 = soup.find_all(class_='a-text-right mojo-field-type-money')
for gr in gr1:
    movie_gross.append(gr.text)
print (movie_gross)
```

```python
movie_year=[]
yr1 = soup.find_all(class_='a-text-left mojo-field-type-year')
for yr in yr1:
    movie_year.append(yr.text)
print (movie_year)
```

```python
import pandas as pd
movieranking = pd.DataFrame({
    "rank": movie_ranks,
    "title": movie_titles,
    "gross": movie_gross[0:600:3],
    "year": movie_year,
             })
print (movieranking)
```

```python
export_csv = movieranking.to_csv ('webscrape_group2.csv', index = None, header=True) 
```
